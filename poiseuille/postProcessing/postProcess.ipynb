```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('pressure.csv')
```

```python
df = data.copy()
df = df.rename(columns={'Column1': 'X', 'Column2': 'Pressure'})
df
```

```python
plt.style.use("dark_background")
plt.figure()
df.plot(x = 'X', y = 'Pressure', kind='line', color='yellow')
plt.title('Pressure along X axis')
plt.xlabel('X')
plt.ylabel('Pressure')
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

```python
u_df = pd.read_csv('velocity.csv')
u_df = u_df.drop(['Column3', 'Column4'], axis=1)
u_df = u_df.rename(columns={'Column1':'X', 'Column2': 'Velocity'})
u_df
```

```python
plt.style.use("dark_background")
plt.figure()
u_df.plot(x = 'X', y = 'Velocity', kind='line', color='yellow')
plt.title('Velocity along X axis')
plt.xlabel('X')
plt.ylabel('Pressure')
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5, color='gray')
plt.show()
```
